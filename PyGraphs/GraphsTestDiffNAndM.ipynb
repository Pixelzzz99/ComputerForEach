```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def getInfoFromFiles(fileName, x_list, y_list, z_list):
    roster = list()
    with open(fileName, 'r') as file:
        lines = [line.rstrip('\n') for line in file]
        for line in lines:
            if(line.isdigit()):
                roster.append(int(line))
        x_list += roster[0::2]
        y_list += roster[1::2]
        roster = lines[2::3]
        for element in roster:
            z_list.append(float(element))
```

```python
x_first_version = list()
y_first_version = list()
z_first_version = list()

x_second_version = list()
y_second_version = list()
z_second_version = list()
```

```python
file1 = 'FirstVersionDiffNAndM.txt'
file2 = 'SecondVersionDiffNAndM.txt'
```

```python
getInfoFromFiles(file1, x_first_version, y_first_version, z_first_version)
getInfoFromFiles(file2, x_second_version, y_second_version, z_second_version)
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")

ax.plot3D(x_first_version, y_first_version, z_first_version, "red")
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")

ax.plot3D(x_second_version, y_second_version, z_second_version, "blue")
```
